```python
from sklearn.model_selection import train_test_split # importing the train test split model
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler # to normalize our data to bring the values in a standard format
sc=StandardScaler()
# converts our data into arrays

# can use lable encoding one hot encoding
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential # it sequentially executes the layers in the network.
from keras.layers import Dense
```

```python
# Initialising the ANN
classifier = Sequential()
```

**What is the difference between he and glorot initialization?
Glorot initialization is recommended for tanh- and sigmoid-based neurons, and he  initialization is recommended for ReLU-based neurons.**

```python
# Adding the input layer and the first hidden layer

#Initializers define the way to set the initial random weights of Keras layers.
classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu',input_dim = 4))

# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu')) # here we have considered bias as 1 and for those 6x1 = 6

classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))

classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))

classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))

classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))

classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))

classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))

classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))

# Adding the output layer
classifier.add(Dense(units= 1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))
```

```python

# Compiling the ANN
#Computes the cross-entropy loss between true labels and predicted labels.
#Use this cross-entropy loss for binary (0 or 1) classification applications.
#Calculates how often predictions equal labels.

classifier.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
model_history=classifier.fit(x_train, y_train,validation_split=0.33, batch_size = 10,  epochs = 100)
```

When 3 layers are used -> accuracy is  - 90 and validation accuracy - 93

When 4 layers are used -> accuracy is  - 90 and validation accuracy - 88

When 5 layers are used -> accuracy is  - 92.99 and validation accuracy - 82.08

When 6 layers are used -> accuracy is  - 90.19 and validation accuracy - 90.57

When 7 layers are used -> accuracy is  - 90 and validation accuracy - 91

When 8 layers are used -> accuracy is  - 93 and validation accuracy - 83
